```python
import os
from datetime import datetime, timedelta
from urllib.request import Request, urlopen
from random import sample 
import time
import calendar
from dateutil.relativedelta import *
```

```python
if not os.path.exists('weather_all'):
    os.mkdir('weather_all')
```

```python
Measurements = ['\"desc\"','\"temp\"','\"templow\"','\"baro\"','\"wind\"','\"wd\"','\"hum\"']
```

```python
Cities = ['japan/sapporo','japan/niigata','japan/aomori','japan/kanazawa','japan/hiroshima','japan/sendai',
          'japan/kyoto', 'japan/tokyo', 'japan/fukuoka', 'japan/shizuoka','japan/matsuyama','japan/osaka',
          'japan/nagoya','japan/nagasaki','japan/kagoshima','japan/naha','usa/washington-dc','canada/toronto']
```

```python
begin_date = '2009-07-01'
end_date = '2019-07-01'
```

```python
def scrape_weather(begin_date, end_date, Measurements, city):
    
    current_date = datetime.strptime(begin_date, "%Y-%m-%d")
    end_date = datetime.strptime(end_date, "%Y-%m-%d")+relativedelta(months=+1)#timedelta(month=1) #add one to make loop end on the end date
    
    #https://www.almanac.com/weather/history/DC/Washington/1951-11-04
    lookup_URL = 'https://www.timeanddate.com/weather/{}/historic?month={}&year={}'
    
    #String to store data
    str_data = ''
    
    #Cycle through every date to extract .html
    while current_date != end_date:
        
        #print(current_date.month)
        #print(current_date.year)
        
        formatted_lookup_URL = lookup_URL.format(city,
                                                 current_date.month,
                                                 current_date.year)
        str_month = current_date.strftime("%B")
        
        print(formatted_lookup_URL)
        
        #print(str_month)
        
        try:
            req = Request(formatted_lookup_URL, headers={'User-Agent': 'Mozilla/5.0'})
            str_html = urlopen(req).read().decode('utf-8')
            
            #print(str_html)
            
            for i in range(1,calendar.monthrange(current_date.year,current_date.month)[1]+1):
                index = str_html.find(str_month + ' ' + str(i) + ',')
                str_html = str_html[index:]
                
                str_data += current_date.strftime("%Y-%m") + '-' + str(i)
                
                #for each day, there are 4 time periods
                for j in range(4):
                    for measurement in Measurements:
                        index = str_html.find(measurement)
                        str_html = str_html[index:]
                        index = str_html.find(':')
                        str_html = str_html[index+1:]
                        if measurement == '\"hum\"' :
                            index = str_html.find('}')
                        else:
                            index = str_html.find(',')
                        str_data += ',' + str_html[:index]
                str_data+='\n'

            current_date += relativedelta(months=+1)#timedelta(month=1)  
            
        except:
            print('ERROR\n')
            print(str_data)
            current_date = current_date
            
            UA = sample(UAs,1)
            
            #Wait, too many requests were made
            time.sleep(1)
          
    out_file_name = 'weather_all/' + city[city.find('/')+1:] + '.csv'
    with open(out_file_name, 'w') as out_file:
        out_file.write(str_data)
```

```python
for city in Cities:
    start = time.time()
    scrape_weather(begin_date, end_date, Measurements,city)
    elapsed = time.time()-start
    print('elapsed time: ' + str(elapsed))
```
